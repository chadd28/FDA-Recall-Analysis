```python
import requests
import pandas as pd
```

```python
base = 'https://api.fda.gov/drug/event.json'
fieldTerm = 'patient.reaction.reactionmeddrapt:"fatigue"'
limit = "3"
url = f'{base}?search={fieldTerm}&limit={limit}'

response = requests.get(url)

print("Status Code: ",response.status_code)

print(response.json())
```

```python
data = response.json()
df = pd.json_normalize(data['results'])
df
```
